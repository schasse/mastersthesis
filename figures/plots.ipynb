```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(rc={'figure.figsize':(7,7)})
sns.set_style('darkgrid', {'font.sans-serif': ['Source Code Pro']})
one = '#00527aff'
two = '#c18c00ff'
three = '#c12800ff'
four = '#3494c1ff'
five = '#681500ff'
```

# deploy din

```python
df = pd.DataFrame([
    ["1.", 11, 12],
    ["2.", 8, 9],
    ["3.", 9, 10]
], columns=['exec', '1', '2'])
```

```python
sns.barplot(x="exec", y="2", data=df, color=three, label='kubernetes(9-running)')
ax = sns.barplot(x="exec", y="1", data=df, color=one, label='deployer(6-8)')
ax.set(xlabel='execution', ylabel='seconds')
ax.legend()
```

# whole pipeline

```python
df = pd.DataFrame([["1.", 4, 241, 308, 344, 414], ["2.", 6, 100, 107, 127, 172], ["3.", 4, 91, 99, 117, 192]], columns=['exec', '1', '2', '3', '4', '5'])
```

```python
sns.barplot(x="exec", y="5", data=df, color=three, label='kubernetes(9-running)')
sns.barplot(x="exec", y="4", data=df, color=one, label='deployer(6-8)')
sns.barplot(x="exec", y="3", data=df, color=two, label='save images(4)')
sns.barplot(x="exec", y="2", data=df, color=four, label='build(3)')
ax = sns.barplot(x="exec", y="1", data=df, color=five, label='commit(1-2)')
ax.set(xlabel='execution', ylabel='seconds')
ax.legend()
```
